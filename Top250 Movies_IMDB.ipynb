```python
from selenium import webdriver
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
driver = webdriver.Chrome('chromedriver',chrome_options=chrome_options)
```

```python
browser = webdriver.Chrome('chromedriver',chrome_options=chrome_options)
browser.maximize_window()


web_url = "https://www.imdb.com/chart/top/"  
browser.get(web_url)
time.sleep(1)

movie_list_location = browser.find_element_by_xpath("//tbody[@class='lister-list']")
movie_names = movie_list_location.find_elements_by_xpath("//td[@class='titleColumn']")
movie_ratings = movie_list_location.find_elements_by_xpath("//td[@class='ratingColumn imdbRating']")

movie_name = [name.text for name in movie_names]
movie_link = []
imdb_rating = [rating.text for rating in movie_ratings]


for movie in movie_names:
    link = movie.find_element_by_tag_name("a").get_attribute('href')
    movie_link.append(link)
```

```python
review_link = []
for i in movie_link:
    web_url=i
    browser = webdriver.Chrome('chromedriver',chrome_options=chrome_options)
    browser.get(web_url)
    browser.maximize_window()
    time.sleep(1)

    review_list_location = browser.find_element_by_xpath("//section[@Data-testid='UserReviews']")
    review_names = review_list_location.find_elements_by_xpath("//div[@Data-testid='reviews-header']")
    review_name = [name.text for name in review_names]
    
    for movie in review_names:
        link = movie.find_element_by_tag_name("a").get_attribute('href')
        review_link.append(link)
```

```python
title_name = []
rating = []
review = []
votes = []
review_date = []
users=[]
```

```python
import re
from selenium.webdriver.chrome.options import Options
#importing required Libraries
import pandas as pd   #to create dataframe
import requests       #to send the request to the URL
from bs4 import BeautifulSoup #to get the content in the form of HTML
import numpy as np  # to count the values (in our case)
import click
```

```python
for i in  tqdm(review_link):
    url =i
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    movie_data = soup.findAll('div', attrs= {'class': 'lister-list'})
    review_data = soup.findAll('div', attrs= {'class': 'review-container'})
    name = soup.findAll('a', class_ = 'title')
    rate = soup.findAll('span', class_ = 'rating-other-user-rating')
    reviewing = soup.findAll('div', attrs = {'class': 'text show-more__control'})
    voting = soup.findAll('div', class_ = 'actions text-muted')
    reviewing_date = soup.findAll('span', class_ = 'review-date')
    user_name=soup.findAll('div', class_ = 'display-name-date')
    for r in rate:
        rating.append(r.text.replace('\n', ''))
    for t in name:
        title_name.append(t.text.replace('\n',''))
    for re in reviewing:
        review.append(re.text.replace('\n', ''))
    for v in voting:
        votes.append(v.text.replace('\n', ''))
    for d in reviewing_date:
        review_date.append(d.text.replace('\n', ''))
    for i in user_name:
      users.append(i.text.replace('\n',''))
```

```python
#creating a dataframe using pandas library
a = {'Title': title_name, 'Rating': rating,  'Reviews': review, 'Votes': votes, 'Review Date': review_date}
df = pd.DataFrame.from_dict(a, orient='index')
df = df.transpose()
df
#df.to_excel('WebScrap_250.xlsx' ,index = False )
```
